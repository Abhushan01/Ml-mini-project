**Before Cleaning the dataset**

**Data-Cleaning**

We will first remove useless columns from our dataset

*Note*: The command written below will not give any output, but it will show error when you run this command, this is due to the reason that the column "Home Address" has been deleted from the dataset, when I executed the command first time, the error is due to the fact that column we want to delete no longer exits in the dataset

```
#dropping useless columns
dropping=['Home Address']
df.drop(dropping, axis=1, inplace=True)
```

This command will fill the empty entries with the mean values of their columns

```
df.fillna(df.mean(), inplace=True)
```

Now, we will check how our data- cleaning goes, as you can observe there is no longer any null entry in any column, we have filled those entries with the mean values

```
print(df.info(),"\n")
print(df.isnull().sum())
```
